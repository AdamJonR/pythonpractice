```python
import wordcount
```

```python
# start the processing, accepting the name of the file to open
def main(file_name):
    # variable for storing string contents of file
    string_contents = ''
    
    # use the with statement to elegantly handle automatically closing the file when done
    with open(file_name,"r") as file:
        # store the contents of the file as one big string
        string_contents = file.read()

    # call our words list function to get a list of all of the words in the string from the file
    word_list = wordcount.get_words_list(string_contents)
    # call our counts function to get a tally of the word counts in the string from the file
    word_counts = wordcount.get_counts(word_list)
    
    # loop through the counts and print them
    for word, count in word_counts.items():
        print(f"{word}: {count}")

# everything starts here
main("some_text.txt")
```
